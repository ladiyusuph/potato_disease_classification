```python
#Creating the path object
disease_type = 'earlyblight','lateblight','healthy'
path = Path('/home/ladiyusuph/Ladi/Kaggle/potato_disease_classification/Potato_img')
```

```python
#Loading the image files
fns = get_image_files(path)
fns
```

```python
#Checking for failed downloads
failed = verify_images(fns)
failed
```

## Model Building

```python
#Creating a data block
disease = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
```

```python
#Loading the data
dls = disease.dataloaders(path)
```

```python
#Viewing the loaded images
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
#Augmenting the data to help improve accuracy in the wild
disease = disease.new(item_tfms=RandomResizedCrop(128, min_scale=0.3))
dls = disease.dataloaders(path)
dls.train.show_batch(max_n=4, nrows=1, unique=True)
```

```python
#Training a base model
learn = vision_learner(dls, resnet18, metrics=[error_rate,accuracy])
learn.fine_tune(5)
```

```python
#Plotting a confusion matrix to view the model performance
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
#Plotting the top five losses in the model
interp.plot_top_losses(5, nrows=1)
```

```python
#Cleaning the data from the base model
from fastai.vision.widgets import * 

cleaner = ImageClassifierCleaner(learn)
cleaner
```

```python
#Deleting the removed data 
for idx in cleaner.delete(): cleaner.fns[idx].unlink()

#Transferring the relabeled data
import os

for idx, cat in cleaner.change():
    f = cleaner.fns[idx]
    os.rename(f, f.parent/(f.parent.name+f.stem+f.suffix))
    shutil.move(str(f.parent/(f.parent.name+f.stem+f.suffix)), path/cat)    
```

```python
#Exporting the trained model
learn.export()
     
path = Path()
path.ls(file_exts='.pkl')
```

## Model Deployment

```python
#Loading the exported Model
learn_inf = load_learner(path/'export.pkl')
```

```python
#Predicting with the loaded model
learn_inf.predict('earlyblight.jpg')
```

### Building a Notebook app

```python
#Asking for user upload
btn_upload = widgets.FileUpload()
btn_upload
```

```python
#Viewing the user input
import matplotlib.pyplot as plt
from PIL import Image
from io import BytesIO
img = Image.open(BytesIO(btn_upload.value[-1].content))
plt.imshow(img)
```

```python
#Converting and resizing user input
out_pl = widgets.Output()
out_pl.clear_output()
with out_pl: display(img.to_thumb(128,128))
out_pl
```

```python
#Converting input to array and making prediction
img = np.array(img)
pred,pred_idx,probs = learn_inf.predict(img)
```

```python
#Creating the interface in the notebook

#Adding a label widget
lbl_pred = widgets.Label()
lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

#Adding a classify button
btn_run = widgets.Button(description='Classify')

#Creating the label fuction
def on_click_classify(change):
    img = Image.open(BytesIO(btn_upload.value[-1].content))  # change this
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    img = np.array(img)                                      # and add this
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)

#Putting back btn_upload to a widget for next cell
btn_upload = widgets.FileUpload()
```
